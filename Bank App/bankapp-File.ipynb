```python
import json
import sys
from getpass import getpass
from datetime import datetime
import time
```

```python
bank = {'acc':[],
        'user':[],
        'pass':[],
         'bal':[]}
user= []
firstname = []
lastname = []
adm='umang'
adm_ps='umang'
file = open("bank.db",'w')
log = open("log.txt",'w+')
```

```python

def bankuse():
    print("\n\t1.Login\n\t2.Signup\n\t3.Admin\n\t4.Exit")
    choice = input("\n\tenter your choice: ")
    if choice=='1':
        login()
    
    elif choice=='2':
        signup()
    
    elif choice=='3':
        admin()
    
    elif choice == '4':
        ch1= input("\n\tDo You really want to exit??")
        if ch1=="y" or ch1=="yes":
            print("\n\tThank you for using Services")
            log.close()
            file.close()
        else:
            bankuse()
    else:
        print("\n\tInvalid Choice")

def login():
    loginnum = int(input("\n\tEnter your Account number: "))
    if loginnum in bank['acc']:
        index = bank['acc'].index(loginnum)
        passwrd2 = getpass("\n\tEnter your password-->") 
        if passwrd2 == bank['pass'][index]:
            print("\n\tWelcome !!")
            print("\n\t1.Debit\n\t2.Credit\n\t3.Check Balance\n\t4.Change Password\n\t5.Log Out")
            
            loginchoice(index,loginnum)
            
        else:
            print("\n\tinvalid password")
            return
    else:
        print("\n\tinvalid acc number")
        return
def loginchoice(index,loginnum):
    logchoice = input("\n\tEnter your choice: ")
    if logchoice=='1':
                    debitamount=float(input("\n\tEnter the amount you want to Debit: "))
                    debit(debitamount,index,loginnum)
    elif logchoice=='2':
                    creditamount = float(input("\n\tenter the Amount you want to Credit: "))
                    credit(creditamount,index,loginnum)
    elif logchoice=='3':
                    checkbalance(index,loginnum)
    elif logchoice=='4':
                    newpass = getpass("\n\tPlease enter New Password: ")
                    changepassword(newpass,index,loginnum)
    elif logchoice =='5':
                    logout = input("\n\tDo you want to logout? (y/n) ")
                    if logout.lower().startswith("n"):
                        loginchoice(index,loginnum)
                    else:
                        print("\n\tYou are logged out")
                        file.close()
                        log.close()
                        return
                        
def credit(creditamount,index,loginnum):
    
    bank['bal'][index] = creditamount+bank['bal'][index]
    print(f"\n\tYour Balance is {bank['bal'][index]}")
    log.write(f"{loginnum} credited {creditamount} at {time.ctime()}\n")
    json.dump(bank,file)
    loginchoice(index,loginnum)
       
def debit(debitamount,index,loginnum):
    amount= bank['bal'][index]-debitamount 
    if amount>0: 
        bank['bal'][index]=amount
        print(f"\n\tYour Balance is {amount}")
    elif amount==0:
        ask=input("\n\tYour balance will be zero. Do you want to proceed? (y/n) --> ")
        if ask=='y':
            bank['bal'][index]=debitamount-bank['bal'][index]
            print("\n\tYour balance is 0")
        else:
            print("\n\tChoose other options")
    else:
        print("\n\tYou don't have enough balance")
    log.write(f"{loginnum} debited {amount} at {time.ctime()}\n")
    json.dump(bank,file)
    loginchoice(index,loginnum)
def checkbalance(index,loginnum):
    print(f"\n\tYour Balance is {bank['bal'][index]}")
    log.write(f"{loginnum} checked password at {time.ctime()}\n")
    loginchoice(index,loginnum)
def changepassword(newpass,index,loginnum):
    bank['pass'][index]=newpass
    print("\n\tYour password is updated")  
    log.write(f"{loginnum} changed password at {time.ctime()}\n")
    json.dump(bank,file)
    loginchoice(index,loginnum)
    
                
def signup():
    fname= input("\n\tenter your first name: ")
    lname=input("\n\tenter  your last name: ")
    suppassword = getpass("\n\tEnter your password: ")
    firstname.append(fname)
    lastname.append(lname)
    if bank['acc']==[]:
        bankaccnumber = 1000# bank['acc'].append(1000)
        bank['acc'].append(bankaccnumber)
    else:
        bankaccnumber = bank['acc'][-1]+1
        bank['acc'].append(bankaccnumber)
    user.append([fname,lname])
    username = [i+"@"+j for i,j in zip(fname,lname)]
    bank['user'].append([fname,lname])
    bank['pass'].append(suppassword)
    bank['bal'].append(0)
    json.dump(bank,file)
    print(f"\n\t Welcome ! your account number is {bankaccnumber}. You  may login now !!")
    log.write(f"{bankaccnumber} signed up at {time.ctime()}\n")
    ask= input("\n\tDo you want to login (y/n)? ")
    if ask=='y':
        login() 
    else:
        print("\n\tThank You for Opening your account. Please visit again !!")
        bankuse()
    
    
    
def admin():
    count = 0
    while count<3:
        adminuser = input("\n\tPlease enter Admin Username: ")
        adminpass = getpass("\n\tEnter Admin Password: ")
        if adminuser== adm and adminpass==adm_ps:
            log.write(f"{adminuser} logged in at {time.ctime()}\n")
            
            print("\n\tAccount Information: \n")
            for var in bank:
                print("\n\t",var,"=",bank[var])
            break
        else:
            print("\n\tAccess Denied")
            count+=1
            if count==3:
                  print("\n\tYou have used all 3 attempts. Please check your credentials !")
            else:
                print(f"\n\tYou have used {count} attempts. You have {3-count} attempts left")
                log.write(f"3 unsuccessful login attempt of {adminuser} at {time.ctime()}\n")
```
